```python
url = 'https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteorologiczne/dobowe/synop/'
```

```python
import requests
r = requests.get(url)
```

```python
# print(r.text)
```

```python
# LISTA PLIKW

from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'lxml')
# soup.prettify()
web_links = soup.select('a')
# for link in web_links:
#     print(link)
    
actual_web_links = [web_link['href'] for web_link in web_links] 

zip_files = []

for link in actual_web_links:
    new_r = requests.get(url+link)
    new_soup = BeautifulSoup(new_r.text, 'lxml')
    web_links = new_soup.select('a')
#     print(web_links)
    for x in [web_link['href'] for web_link in web_links]:
        if 'zip' in x:
            zip_files.append(url+link+x)
    
print(zip_files)
# print(web_links)
# for el in soup.body.table:
#     print(el)
```

```python
# POBRANIE ZIP

import time
import random
import os

path = 'DANE/'

for file in zip_files[::-1]:
#     time.sleep(random.randint(1,10))
    filename = file.split('/')[-1]
    if not os.path.exists(path+filename):
        r = requests.get(file)
        print("pobieranie: "+filename, zip_files.index(file),'/',len(zip_files))
#         print()
        with open('DANE/'+filename,'wb') as output_file:
            output_file.write(r.content)
```

```python
from zipfile import ZipFile

with ZipFile("DANE/2019_300_s.zip","r") as zip_object:
    zip_object.extractall("DANE/ALL/")

zip_object.close()
```

```python
# ROZPAKOWANIE ZIP

import os

directory = "DANE/"

for file in os.scandir(directory):
    print(str(file.name))
    if file.is_file():
        with ZipFile("DANE/"+file.name,"r") as zip_object:
            zip_object.extractall("DANE/ALL/")
```

```python
import pandas as pd
import glob
import os

files = os.path.join("DANE/ALL/", "s_d_*.csv")
files = glob.glob(files)

# for file in files:
#     df = pd.read_csv(file, encoding='latin1')

combined_csv = pd.concat([pd.read_csv(file, encoding='latin1', header=None) for file in files ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')

# print(df.head())
```
